```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import time
```

```python

def GetFundTimeSeries(fundcode):
    str1 = 'https://coin.zerodha.com/api/fund-info?graph_type=normal&scheme_id='
    fundcode = str(fundcode)
    str2 = '&session_token='

    string =  str1+fundcode+str2


    response = requests.get(string) 
    json_data = json.loads(response.text)

    #Time Series Data
    graphdata = pd.DataFrame.from_dict(json_data['data']['eq_graph'])
    graphdata['x'] = pd.to_datetime(graphdata['x']/1000,unit='s')
    graphdata['x'] = graphdata['x'].dt.date
    graphdata.set_index('x',inplace=True)
    
    #Exit Load, Expense Ratio and other data
    information_df = pd.DataFrame.from_dict(json_data['data']['master'])
    #price returns data
    returns_df = pd.DataFrame.from_dict(json_data['data']['nav_prices']) 

    
    
    
    return graphdata,information_df,returns_df

```
